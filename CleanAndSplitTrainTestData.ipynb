```python
import string
import re
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import csv
```

```python
df = pd.read_csv('DataMerged.csv', sep=',',index_col='ID')
print(df.shape)
df.head()
```

```python
df.drop_duplicates(subset ="comment_content", inplace = True)
df['comment_content'].replace('', np.nan, inplace=True)
df.dropna(subset=['comment_content'], inplace=True)
print(df.shape)
```

```python
df.mean()
```

```python
df.sum(numeric_only=True)
```

```python
list_sum = list(df.sum(numeric_only=True))
idx = [i for i in range(len(list_sum)) if list_sum[i] < 100]
drop_column = [list(df)[i+1] for i in idx]
df = df.drop(drop_column, axis =1)
df.sum(numeric_only=True)
```

```python
drop_list = df[(df['Giao hang'] == 0) & (df['Dong goi'] == 0) 
               & (df['Gia ban'] == 0) & (df['Thiet ke'] == 0)
               & (df['Chat luong san pham'] == 1)
               & (df['Cham soc khach hang'] == 0) & (df['Do nhay'] == 0)]
df = df[~df[:].isin(drop_list)]
df= df.dropna()
df.describe()
```

```python
df.head(20)
```

```python
#Convert all to lower case
df.comment_content = df.comment_content.apply(lambda x: x.lower())
#Remove punctuation, number, Emojify
df.comment_content = df.comment_content.str.replace(r"""[^a-z\s
                                    \à\á\ạ\ả\ã\â\ầ\ấ\ậ\ẩ\ẫ\ă\ằ\ắ\ặ\ẳ\ẵ
                                    \è\é\ẹ\ẻ\ẽ\ê\ề\ế\ệ\ể\ễ
                                    \ì\í\ị\ỉ\ĩ
                                    \ò\ó\ọ\ỏ\õ\ô\ồ\ố\ộ\ổ\ỗ\ơ\ờ\ớ\ợ\ở\ỡ
                                    \ù\ú\ụ\ủ\ũ\ư\ừ\ứ\ự\ử\ữ
                                    \ỳ\ý\ỵ\ỷ\ỹ
                                    \đ]""", '')
df.head()
```

```python
target = list(df)[1:]

X_train, X_test, y_train, y_test = train_test_split(df['comment_content'], df[target[:]], 
                                                    stratify=df['Do nhay'], test_size=0.20, random_state=123456)
y_test.mean()
```

```python
y_train.mean()
```

```python
df_X_test = pd.DataFrame(X_test, columns=['comment_content'])
df_y_test = pd.DataFrame(y_test, columns=target)

col = df_X_test['comment_content']
df_y_test.insert(df_y_test.columns.get_loc('Giao hang'), col.name, col, allow_duplicates=True)
df_y_test.to_csv('Test.csv',index=False,header=True, quoting=csv.QUOTE_NONNUMERIC)
```

```python
df_X_train = pd.DataFrame(X_train, columns=['comment_content'])
df_y_train = pd.DataFrame(y_train, columns=target)

col = df_X_train['comment_content']
df_y_train.insert(df_y_train.columns.get_loc('Giao hang'), col.name, col, allow_duplicates=True)
df_y_train.to_csv('Train.csv',index=False,header=True, quoting=csv.QUOTE_NONNUMERIC)
```
